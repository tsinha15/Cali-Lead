```python
def map_lead_contam(x):
    BLL_mt_2 = x[x["Percent_Under_21_with_>=_4.5BLL"] > 2]
    two_to_four = BLL_mt_2_lt_et_4 = BLL_mt_2[BLL_mt_2["Percent_Under_21_with_>=_4.5BLL"] <= 4]
    zero_to_two = BLL_lt_et_2 = x[x["Percent_Under_21_with_>=_4.5BLL"] <= 2]
    four_and_beyond = BLL_mt_4 = x[x["Percent_Under_21_with_>=_4.5BLL"] > 4]
    global data
    global layout
    global fig
    data = [go.Scattermapbox(
    lat = zero_to_two["Latitude"],
    lon = zero_to_two["Longitude"],
    mode="markers",
    marker=dict(
    size = 5,
    color = "yellow",
    opacity = .8,
    ),
    name="Less than or equal to 2%",
    hoverinfo="lat+lon"
    ),
    go.Scattermapbox(
    lat = two_to_four["Latitude"],
    lon = two_to_four["Longitude"],
    mode="markers",
    marker=dict(
    size = 6,
    color = "orange",
    opacity = .8,
    ),
    name="Between 2% and 4%",
    hoverinfo="lat+lon"
    ),
    go.Scattermapbox(
    lat = four_and_beyond["Latitude"],
    lon = four_and_beyond["Longitude"],
    mode="markers",
    marker=dict(
    size = 7,
    color = "red",
    opacity = .8,
    ),
    name="Greater than 4%",
    hoverinfo="lat+lon")
    ]
    layout = go.Layout(
    title = "California Counties showing Percentages of Under 21 Individuals with BLL's >= 4.5",
    autosize=True,
    hovermode="closest",
    showlegend=True,
    mapbox=dict(
    accesstoken=mapbox_access_token,
    bearing=0,
    pitch=0,
    zoom = 5,
    style="dark",
    center=dict(lat=37.1, lon=-119)
    ),
    width=900,
    height=800
    )
    fig = dict(data=data, layout=layout)
    py.iplot(fig, filename="smtg1")
```
